```python
import pandas as pd
import numpy as np 
```

```python
data={
    'ID':[1,2,3,4,5,6],
    'Name':['Ram','Shyam','Raju','Kaju','Prem','Rahul'],
    'Age':[20,32,23,40,21,30],
    'State':["MH","GJ","UP","MP","MH","UP"],
    'Sales':[200,300,450,320,420,250]
}
```

```python
df=pd.DataFrame(data)
print("Original Data : ")
df
```

```python
# transform data

df['Name_Upper']=df['Name'].str.upper()
df.head()
```

```python
#multicast 
df1=df.copy()
df2=df.copy()

df1['Age']+=5
df2['Sales']*=1.1

print("Df 1 Updated Ages ")
print(df1.head())

print("Df 2 Updated Weights ")
print(df2.head())
```

```python
# condition split
high_sales=df[df['Sales']>300]
low_sales=df[df['Sales']<=300]

print("High sales : " , high_sales)
print("Low sales : ",low_sales)
```

```python
# Aggregation 
agg_df=df.groupby(['State'])['Sales'].sum()
agg_df
```

```python
#Sorting in decending order
sorted_df=df.sort_values(by='Sales',ascending=False)
sorted_df
```

```python
#Derieved Column

df["Sales Category"]=df["Sales"].apply(lambda x : 'High' if x > 300 else 'Low')
df.head()
```
